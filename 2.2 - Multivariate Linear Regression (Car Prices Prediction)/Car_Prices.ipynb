```python
import pandas
from sklearn.linear_model import LinearRegression
```

```python
data_frame = pandas.read_csv('/content/drive/MyDrive/Colab Folders/Car Prices/carprices.csv')
data_frame
```

```python
cars = pandas.get_dummies(data_frame['Car Model'])
cars
```

```python
data_frame = pandas.concat([data_frame, cars], axis='columns')
data_frame = data_frame.drop(['Car Model'], axis='columns')
```

```python
data_frame
```

```python
X = data_frame[['Mileage', 'Age(yrs)', 'Audi A5', 'BMW X5', 'Mercedez Benz C class']]
X
```

```python
y = data_frame[['Sell Price($)']]
y
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.predict([[45000, 4, 0, 0, 1]])
```

```python
model.predict([[86000, 7, 0, 1, 0]])
```

```python
model.score(X, y)
```
