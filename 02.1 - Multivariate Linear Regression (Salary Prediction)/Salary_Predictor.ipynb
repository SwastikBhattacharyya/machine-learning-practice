```python
import pandas
import numpy
from word2number import w2n
from sklearn.linear_model import LinearRegression
```

```python
data_frame = pandas.read_csv('/content/drive/MyDrive/Colab Folders/Salary Predictor/hiring.csv')
data_frame
```

```python
data_frame['experience'] = data_frame['experience'].fillna('zero')
data_frame['experience'] = data_frame['experience'].apply(w2n.word_to_num)
data_frame
```

```python
import math

test_score_mean = math.floor(data_frame['test_score(out of 10)'].mean())
data_frame['test_score(out of 10)'] = data_frame['test_score(out of 10)'].fillna(test_score_mean)
data_frame
```

```python
linear_regression = LinearRegression()
linear_regression.fit(data_frame[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], data_frame['salary($)'])
```

```python
linear_regression.predict([[2, 9, 6], [12, 10, 10]])
```
